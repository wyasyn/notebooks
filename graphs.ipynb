```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [23, 45, 56, 78, 12],
    'Another_Column': [10, 20, 30, 40, 50]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python

# Line Plot
plt.figure(figsize=(8, 5))
sns.lineplot(x='Category', y='Values', data=df, marker='o', label='Values')
sns.lineplot(x='Category', y='Another_Column', data=df, marker='s', label='Another Column')
plt.title('Line Plot Example')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Values', data=df, hue='Category', palette='viridis', legend=False)
plt.title('Bar Plot Example')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Values', y='Another_Column', data=df, hue='Category', palette='deep', s=100)
plt.title('Scatter Plot Example')
plt.xlabel('Values')
plt.ylabel('Another Column')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['Values'], bins=5, kde=True, color='blue')
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df[['Values', 'Another_Column']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap Example')
plt.show()
```
